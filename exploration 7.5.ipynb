Now we try to implement improved euler method with the recursion law given in the text

```python
def improved_euler(f, t0, y0, h, n):
    y = y0
    t = t0
    pts = [(t,y)]
    
    for _ in range(0, n):
        # Predictor step (using Euler method)
        k1 = f(t, y)
        y_predict = y + h * k1
        
        # Corrector step
        k2 = f(t + h, y_predict)
        y = y + h * (k1 + k2) / 2  # Average of slopes
        
        t = t + h
        pts.append((t,y))
    
    return pts
```

```python
#step size= 0.1imeuler
pts = improved_euler(f,0,1,0.1,10)
print(pts)
#Error
error4= pts[-1][1]-soln(pts[-1][0])
print(error4)
```

```python
#step size= 0.05imeuler
pts = improved_euler(f,0,1,0.05,20)
print(pts)
#Error
error5= pts[-1][1]-soln(pts[-1][0])
print(error5)

```

```python
ratio3= error4/error5
print(ratio3)
```

When we go from stepsize 0.1 to 0.05, our error is $\frac{1}{3.85}$ th

Now we implement the Runge-Kutta method.

```python
def runge_kutta(f,t0,y0,h,n):
    y = y0
    t = t0
    pts = [(t,y)]
    for _ in range(0,n):
        k1 = f(t,y)
        k2 = f(t+h/2,y+k1*h/2)
        k3 = f(t+h/2,y+k2*h/2)
        k4 = f(t+h,y+k3*h)
        y = y + (k1+2*k2+2*k3+k4)*h/6
        t = t + h
        pts.append((t,y))
    return pts
```

```python
#step size= 0.1rk
pts = runge_kutta(f,0,1,0.1,10)
print(pts)
#Error
error7= pts[-1][1]-soln(pts[-1][0])
print(error7)
```

```python
#step size= 0.05rk
pts = runge_kutta(f,0,1,0.05,20)
print(pts)
#Error
error8= pts[-1][1]-soln(pts[-1][0])
print(error8)
```

```python
ratio5= error7/error8 
print(ratio5)
```

Now we do the second ODE $x' = 2t(1+x^2)$. Taking the initial condition $x(0)=0$, we get the analytical solution to be $x(t)= tan(t^2)$ 

```python
import numpy as np

def euler(f,t0,y0,h,n):
    y = y0
    t = t0
    pts = [(t,y)]
    for _ in range(0,n):
        y = y + h * f(t,y)
        t = t + h
        pts.append((t,y))
    return pts

def f(t,y):
    return 2*t*(1+y**2)

def soln(t):
    return np.tan(t**2)

# List of step sizes and corresponding n values
step_sizes = [0.1, 0.05, 0.01]
n_values = [10, 20, 100]

errors = []

# Loop to compute and print errors for each step size
for h, n in zip(step_sizes, n_values):
    pts = euler(f,0,0,h,n)
    print(f"Points for step size {h}:")
    print(pts)
    
    error = pts[-1][1] - soln(pts[-1][0])
    errors.append(error)
    
    print(f"Error for step size {h}:", error)
    print()

# Compute and print ratios
for i in range(len(errors)-1):
    ratio = errors[i] / errors[i+1]
    print(f"Ratio {i+1}:", ratio)
```
